```python
import pandas as pd
df_image_features = pd.read_csv('./image_features/df_image_features.csv')
df_image_features
```

```python
# Inspect columns and basic info
print('Shape:', df_image_features.shape)
print('Columns count:', len(df_image_features.columns))
print('First 10 columns:', df_image_features.columns[:10].tolist())
print('Last 10 columns:', df_image_features.columns[-10:].tolist())
# Try to detect id column candidates
possible_id_cols = [c for c in df_image_features.columns if 'pid' in c.lower() or 'nid' in c.lower() or 'combo' in c.lower()]
print('Possible ID columns:', possible_id_cols)
# Show dtypes of first 5 non-numeric columns
non_numeric = [c for c in df_image_features.columns if not pd.api.types.is_numeric_dtype(df_image_features[c])]
print('Non-numeric columns (sample):', non_numeric[:5])
df_image_features.head(2)
```

```python
# Aggregate embeddings per pid_nid_combo using Mean-Max concatenation
ID_COL = 'pid_nid_combo'
# Select embedding columns strictly as digit-only names within expected range
embedding_cols = [c for c in df_image_features.columns if c.isdigit()]
assert len(embedding_cols) == 1152, f"Expected 1152 embedding columns, got {len(embedding_cols)}"
# Group
grp = df_image_features.groupby(ID_COL)[embedding_cols]
mean_df = grp.mean().add_prefix('mean_')
max_df  = grp.max().add_prefix('max_')
# Concatenate (column-wise)
agg_df = pd.concat([mean_df, max_df], axis=1)
# Also provide a pure 0..2303 naming if needed for downstream models
agg_values = agg_df.values
agg_flat_df = pd.DataFrame(agg_values, index=agg_df.index, columns=[f'emb_{i}' for i in range(agg_values.shape[1])])
print('Aggregated shape (with prefixes):', agg_df.shape)
print('Aggregated shape (flat):', agg_flat_df.shape)
# Merge back optional labels (e.g., malignancy_label) by first occurrence
label_cols = [c for c in ['malignancy_label','area_mm2','patient_id'] if c in df_image_features.columns]
if label_cols:
    first_labels = df_image_features.groupby(ID_COL)[label_cols].first()
    agg_df = agg_df.merge(first_labels, left_index=True, right_index=True, how='left')
    agg_flat_df = agg_flat_df.merge(first_labels, left_index=True, right_index=True, how='left')
    print('Added label columns:', label_cols)
# Save
output_path_prefixed = './image_features/df_nodule_embeddings_mean_max_prefixed.csv'
output_path_flat = './image_features/df_nodule_embeddings_mean_max.csv'
agg_df.reset_index().to_csv(output_path_prefixed, index=False)
agg_flat_df.reset_index().to_csv(output_path_flat, index=False)
print('Saved to:', output_path_prefixed, 'and', output_path_flat)
# Preview
agg_df.head(3)
```

```python
# 1. 加载聚合后的肺结节级别 Mean-Max 拼接向量
import os
import pandas as pd
import numpy as np
from sklearn.model_selection import StratifiedKFold
from sklearn.preprocessing import StandardScaler
from sklearn.pipeline import Pipeline
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score, confusion_matrix
import matplotlib.pyplot as plt
import seaborn as sns

# 尝试导入 xgboost，如未安装则安装
try:
    import xgboost as xgb
except ImportError:
    import sys, subprocess
    subprocess.run([sys.executable, '-m', 'pip', 'install', 'xgboost'])
    import xgboost as xgb

AGG_PATH = './image_features/df_nodule_embeddings_mean_max.csv'  # 使用平铺版本 emb_0..emb_2303
assert os.path.exists(AGG_PATH), f'聚合文件未找到: {AGG_PATH}'
df_nodule = pd.read_csv(AGG_PATH)
print('Loaded shape:', df_nodule.shape)
print('Columns (sample):', df_nodule.columns[:10].tolist())
# 确认标签列
assert 'malignancy_label' in df_nodule.columns, '缺少 malignancy_label 标签列'
df_nodule.head(3)
```

```python
# 2. 准备特征与标签
ID_COL = 'pid_nid_combo'
LABEL_COL = 'malignancy_label'
feature_cols = [c for c in df_nodule.columns if c.startswith('emb_')]
X = df_nodule[feature_cols].values
y = df_nodule[LABEL_COL].values
print('Feature matrix shape:', X.shape)
print('Label distribution:', {int(k): int(v) for k,v in pd.Series(y).value_counts().items()})
```

```python
# 3. 定义 5 折评估函数（StratifiedKFold）
from typing import Tuple, Dict

def evaluate_model(model, X, y, scaler=True, random_state=42) -> Dict:
    skf = StratifiedKFold(n_splits=5, shuffle=True, random_state=random_state)
    accuracies = []
    all_true, all_pred = [], []
    for fold, (train_idx, test_idx) in enumerate(skf.split(X, y), 1):
        X_train, X_test = X[train_idx], X[test_idx]
        y_train, y_test = y[train_idx], y[test_idx]
        if scaler:
            pipe = Pipeline([
                ('scaler', StandardScaler()),
                ('clf', model)
            ])
        else:
            pipe = Pipeline([
                ('clf', model)
            ])
        pipe.fit(X_train, y_train)
        y_pred = pipe.predict(X_test)
        acc = accuracy_score(y_test, y_pred)
        accuracies.append(acc)
        all_true.extend(y_test)
        all_pred.extend(y_pred)
        print(f'Fold {fold} accuracy: {acc:.4f}')
    mean_acc = np.mean(accuracies)
    cm = confusion_matrix(all_true, all_pred)
    return {'mean_accuracy': mean_acc, 'confusion_matrix': cm, 'accuracies': accuracies}

print('评估函数已定义。')
```

```python
# 4. 逻辑回归评估
log_reg = LogisticRegression(max_iter=2000, class_weight='balanced', random_state=42, solver='lbfgs')
results_log = evaluate_model(log_reg, X, y, scaler=True)
print('Mean accuracy (LogReg):', f"{results_log['mean_accuracy']:.4f}")
cm_log = results_log['confusion_matrix']
plt.figure(figsize=(4,3))
sns.heatmap(cm_log, annot=True, fmt='d', cmap='Blues')
plt.title('Confusion Matrix - Logistic Regression')
plt.xlabel('Predicted')
plt.ylabel('True')
plt.tight_layout()
plt.show()
```

```python
# 5. SVM 评估 (使用线性核，适合高维稀疏/致密向量)
svm_clf = SVC(kernel='linear', class_weight='balanced', random_state=42)
results_svm = evaluate_model(svm_clf, X, y, scaler=True)
print('Mean accuracy (SVM):', f"{results_svm['mean_accuracy']:.4f}")
cm_svm = results_svm['confusion_matrix']
plt.figure(figsize=(4,3))
sns.heatmap(cm_svm, annot=True, fmt='d', cmap='Purples')
plt.title('Confusion Matrix - SVM (Linear)')
plt.xlabel('Predicted')
plt.ylabel('True')
plt.tight_layout()
plt.show()
```

```python
# 6. XGBoost 评估 (适合非线性关系)
xgb_clf = xgb.XGBClassifier(
    n_estimators=300,
    max_depth=6,
    learning_rate=0.05,
    subsample=0.9,
    colsample_bytree=0.9,
    reg_lambda=1.0,
    reg_alpha=0.0,
    objective='binary:logistic',
    eval_metric='logloss',
    random_state=42,
    n_jobs=-1
)
# XGBoost 内部可处理未缩放特征，这里仍使用 scaler=True 做统一比较
results_xgb = evaluate_model(xgb_clf, X, y, scaler=True)
print('Mean accuracy (XGBoost):', f"{results_xgb['mean_accuracy']:.4f}")
cm_xgb = results_xgb['confusion_matrix']
plt.figure(figsize=(4,3))
sns.heatmap(cm_xgb, annot=True, fmt='d', cmap='Greens')
plt.title('Confusion Matrix - XGBoost')
plt.xlabel('Predicted')
plt.ylabel('True')
plt.tight_layout()
plt.show()
```

```python
# 7. 汇总三个模型的平均准确率与混淆矩阵
summary = pd.DataFrame({
    'Model': ['LogisticRegression','SVM(Linear)','XGBoost'],
    'Mean_Accuracy': [results_log['mean_accuracy'], results_svm['mean_accuracy'], results_xgb['mean_accuracy']],
    'Fold_Accuracies': [results_log['accuracies'], results_svm['accuracies'], results_xgb['accuracies']]
})
print(summary)

# 条形图比较
plt.figure(figsize=(5,3))
sns.barplot(data=summary, x='Model', y='Mean_Accuracy', palette='Set2', hue='Model')
plt.ylim(0.6, 0.8)
plt.title('Mean Accuracy (5-fold) Comparison')
plt.ylabel('Mean Accuracy')
plt.tight_layout()
# Show number on bars
for i, row in summary.iterrows():
    plt.text(i, row['Mean_Accuracy'] + 0.005, f"{row['Mean_Accuracy']:.3f}", ha='center')
plt.show()
```
