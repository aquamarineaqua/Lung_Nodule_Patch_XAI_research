```python
from sklearn.linear_model import LogisticRegressionCV
import numpy as np

# --- 1. Prepare data (keep your code) ---
X_cols = [col for col in df_nodule_features.columns if "max" in col or "mean" in col]
y_col = 'malignancy_label'
X = df_nodule_features[X_cols]
y = df_nodule_features[y_col]

# --- 2. Standardization (LASSO is very sensitive to scale, must standardize) ---
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# --- 3. Build LASSO model (with automatic hyperparameter tuning) ---
# Cs=10: Try 10 different regularization strengths
# penalty='l1': Enable LASSO mode
# solver='saga': Optimizer supporting L1, suitable for larger datasets
lasso_model = LogisticRegressionCV(
    Cs=np.logspace(-2, 2, 20),  # Search range for C (0.01 to 100)
    cv=5,                       # Inner 5-fold to find optimal C
    penalty='l1',
    solver='saga',              # Or 'liblinear'
    max_iter=10000,             # LASSO converges slowly, need more iterations
    scoring='roc_auc',
    random_state=42,
    n_jobs=-1                   # Parallel computation
)

# --- 4. Train and view results ---
# Note: Fit once here to see which features are selected and the best AUC
lasso_model.fit(X_scaled, y)

print(f"Best C (Regularization Strength): {lasso_model.C_[0]}")
print(f"Model AUC on training data: {lasso_model.score(X_scaled, y)}")

# --- 5. Proper 5-fold cross-validation evaluation (outer CV) ---
# For rigorous performance evaluation, still use cross_val_score
cv_scores = cross_val_score(lasso_model, X_scaled, y, cv=5, scoring='roc_auc')
print("LASSO 5-Fold CV AUC:", cv_scores)
print("LASSO Mean AUC:", cv_scores.mean())

# --- 6. [Core Value] View selected features ---
# Features with coef_ not equal to 0 are selected by LASSO
coefs = pd.Series(lasso_model.coef_[0], index=X_cols)
selected_features = coefs[coefs != 0].sort_values(ascending=False)

print("\n=== Important Features Selected by LASSO (coefficients != 0) ===")
print(selected_features)
print(f"\nOriginal features: {len(X_cols)}, After selection: {len(selected_features)}")
```

# 2 Statistical Significance Testing for Features

Use the following methods to evaluate the statistical significance of features:
1. **statsmodels Logistic Regression** - Obtain p-values, confidence intervals, and Wald statistics for each feature
2. **Bootstrap** - Evaluate the stability of coefficient estimates

```python
import statsmodels.api as sm

# --- Method 1: Use statsmodels to get complete statistical information ---
# Note: statsmodels logistic regression has no regularization by default, so we analyze features selected by LASSO

# Get features selected by LASSO
selected_feature_names = selected_features.index.tolist()
print(f"LASSO selected {len(selected_feature_names)} features, performing statistical tests on these features:\n")

# Use selected features
X_selected = df_nodule_features[selected_feature_names]
X_selected_scaled = scaler.fit_transform(X_selected)

# Add constant term (intercept)
X_with_const = sm.add_constant(X_selected_scaled)

# Fit logistic regression model
logit_model = sm.Logit(y, X_with_const)
result = logit_model.fit(disp=0)  # disp=0 hides iteration process

# Print complete statistical summary
print(result.summary())

# Extract key statistics to DataFrame
stats_df = pd.DataFrame({
    'Feature Name': ['Intercept(const)'] + selected_feature_names,
    'Coef': result.params.values,
    'Std Err': result.bse.values,
    'z-statistic': result.tvalues.values,
    'p-value': result.pvalues.values,
    '95% CI Lower': result.conf_int()[0].values,
    '95% CI Upper': result.conf_int()[1].values
})

# Add significance markers
def significance_stars(p):
    if p < 0.001:
        return '***'
    elif p < 0.01:
        return '**'
    elif p < 0.05:
        return '*'
    elif p < 0.1:
        return '.'
    else:
        return ''

stats_df['Significance'] = stats_df['p-value'].apply(significance_stars)
print("\n" + "="*80)
print("Feature Statistical Significance Summary Table (*** p<0.001, ** p<0.01, * p<0.05, . p<0.1)")
print("="*80)
display(stats_df.round(4))
```

```python
# --- Method 2: Bootstrap to evaluate coefficient stability ---
from sklearn.utils import resample

n_bootstrap = 1000
n_samples = len(y)

# Store coefficients from each bootstrap iteration
bootstrap_coefs = np.zeros((n_bootstrap, len(selected_feature_names)))

print(f"Performing {n_bootstrap} Bootstrap resampling iterations...")

for i in range(n_bootstrap):
    # Resample with replacement
    X_boot, y_boot = resample(X_selected_scaled, y, random_state=i)
    
    # Fit logistic regression (without regularization to evaluate coefficients)
    model_boot = LogisticRegression(max_iter=1000, random_state=42, penalty=None)
    model_boot.fit(X_boot, y_boot)
    bootstrap_coefs[i, :] = model_boot.coef_[0]

# Calculate Bootstrap confidence intervals
ci_lower = np.percentile(bootstrap_coefs, 2.5, axis=0)
ci_upper = np.percentile(bootstrap_coefs, 97.5, axis=0)
coef_mean = np.mean(bootstrap_coefs, axis=0)
coef_std = np.std(bootstrap_coefs, axis=0)

# Determine if confidence interval excludes 0 (excluding 0 means significant)
is_significant = (ci_lower > 0) | (ci_upper < 0)

bootstrap_df = pd.DataFrame({
    'Feature Name': selected_feature_names,
    'Coef Mean': coef_mean,
    'Coef Std': coef_std,
    '95% CI Lower': ci_lower,
    '95% CI Upper': ci_upper,
    'CI excludes 0 (Significant)': is_significant
})

print("\n" + "="*80)
print("Bootstrap Confidence Interval Analysis (95% CI excluding 0 indicates statistical significance)")
print("="*80)
display(bootstrap_df.round(4))

# Significant features summary
significant_features = bootstrap_df[bootstrap_df['CI excludes 0 (Significant)'] == True]['Feature Name'].tolist()
print(f"\nâœ“ Bootstrap significant features ({len(significant_features)}): {significant_features}")
```

```python
# --- Visualization: Feature coefficients and confidence intervals ---
import matplotlib.pyplot as plt
# Set font to support display (Use default English fonts)
plt.rcParams['font.sans-serif'] = ['DejaVu Sans']
plt.rcParams['axes.unicode_minus'] = False  # Fix minus sign display issue

fig, axes = plt.subplots(1, 2, figsize=(16, max(6, len(selected_feature_names)*0.4)))

# Figure 1: statsmodels p-value forest plot
ax1 = axes[0]
colors = ['green' if p < 0.05 else 'gray' for p in stats_df['p-value'][1:]]  # Exclude intercept
y_pos = range(len(selected_feature_names))

ax1.barh(y_pos, stats_df['Coef'][1:], color=colors, alpha=0.7)
ax1.errorbar(stats_df['Coef'][1:], y_pos, 
             xerr=[stats_df['Coef'][1:] - stats_df['95% CI Lower'][1:],
                   stats_df['95% CI Upper'][1:] - stats_df['Coef'][1:]], 
             fmt='none', color='black', capsize=3)
ax1.axvline(x=0, color='red', linestyle='--', linewidth=1)
ax1.set_yticks(y_pos)
ax1.set_yticklabels(selected_feature_names)
ax1.set_xlabel('Coefficient Value')
ax1.set_title('Statsmodels: Feature Coefficients (Green=p<0.05 Significant)')

# Figure 2: Bootstrap confidence intervals
ax2 = axes[1]
colors2 = ['green' if sig else 'gray' for sig in is_significant]
ax2.barh(y_pos, coef_mean, color=colors2, alpha=0.7)
ax2.errorbar(coef_mean, y_pos, xerr=[coef_mean - ci_lower, ci_upper - coef_mean], 
             fmt='none', color='black', capsize=3)
ax2.axvline(x=0, color='red', linestyle='--', linewidth=1)
ax2.set_yticks(y_pos)
ax2.set_yticklabels(selected_feature_names)
ax2.set_xlabel('Coefficient Value')
ax2.set_title('Bootstrap: Feature Coefficients (Green=95%CI excludes 0)')

plt.tight_layout()
plt.show()

# Summary of results from both methods
print("\n" + "="*80)
print("Statistical Significance Summary")
print("="*80)
statsmodels_sig = stats_df[stats_df['p-value'] < 0.05]['Feature Name'].tolist()
statsmodels_sig = [f for f in statsmodels_sig if f != 'Intercept(const)']
print(f"Statsmodels (p<0.05): {statsmodels_sig}")
print(f"Bootstrap (95%CI excludes 0): {significant_features}")
both_sig = set(statsmodels_sig) & set(significant_features)
print(f"\nâœ“ Features significant in both methods: {list(both_sig)}")
```

```python
from PIL import Image
import seaborn as sns
import os

# ========== Re-prepare scaler (for LASSO model's 196-dimensional features) ==========
scaler_full = StandardScaler()
X_full = df_nodule_features[X_cols]
scaler_full.fit(X_full)

# ========== Case Study: Single nodule visualization ==========

def case_study_nodule(nodule_id, df_image_features_withmeta, df_nodule_features, 
                       lasso_model, scaler_full, X_cols, sig_feature_names, sig_01):
    """
    Perform case study visualization for a single nodule
    
    Parameters:
    -----------
    nodule_id : str, Nodule ID (pid_nid_combo)
    """
    
    # 1. Get all patch information for this nodule
    nodule_patches = df_image_features_withmeta[
        df_image_features_withmeta['pid_nid_combo'] == nodule_id
    ]
    
    if len(nodule_patches) == 0:
        print(f"Nodule not found: {nodule_id}")
        return
    
    # 2. Get nodule-level features and labels
    nodule_row = df_nodule_features[df_nodule_features['pid_nid_combo'] == nodule_id]
    true_label = nodule_row['malignancy_label'].values[0]
    
    # 3. Predict using LASSO model
    X_nodule = nodule_row[X_cols].values
    X_nodule_scaled = scaler_full.transform(X_nodule)
    pred_prob = lasso_model.predict_proba(X_nodule_scaled)[0, 1]
    pred_label = lasso_model.predict(X_nodule_scaled)[0]
    
    # 4. Extract values of 12 significant features
    sig_values = nodule_row[sig_feature_names].values[0]
    
    # ========== Visualization ==========
    n_patches = len(nodule_patches)
    
    # Create figure layout
    fig = plt.figure(figsize=(16, 10))
    
    # Title
    label_text = "Malignant" if true_label == 1 else "Benign"
    pred_text = "Malignant" if pred_label == 1 else "Benign"
    correct = "âœ“" if pred_label == true_label else "âœ—"
    
    fig.suptitle(f'Case Study: {nodule_id}\n'
                 f'True Label: {label_text} | Prediction: {pred_text} (Probability: {pred_prob:.3f}) {correct}', 
                 fontsize=14, fontweight='bold')
    
    # Upper half: Display all patch images
    gs = fig.add_gridspec(2, 1, height_ratios=[1, 1.5], hspace=0.3)
    gs_top = gs[0].subgridspec(1, min(n_patches, 5))  # Display up to 5 patches
    
    for i, (idx, row) in enumerate(nodule_patches.iterrows()):
        if i >= 5:  # Maximum 5 patches
            break
        ax = fig.add_subplot(gs_top[i])
        
        # Read image
        img_path = row['img_path_curated']
        if os.path.exists(img_path):
            img = Image.open(img_path)
            ax.imshow(img, cmap='gray')
        else:
            ax.text(0.5, 0.5, 'Image\nNot Found', ha='center', va='center', transform=ax.transAxes)
        
        ax.set_title(f'Patch {i+1}\nArea: {row["area_mm2"]:.1f} mmÂ²', fontsize=10)
        ax.axis('off')
    
    # Lower half: Heatmap of 12 significant features
    gs_bottom = gs[1].subgridspec(1, 2, width_ratios=[2, 1], wspace=0.3)
    
    # Left: Heatmap
    ax_heatmap = fig.add_subplot(gs_bottom[0])
    
    # Prepare heatmap data - separate positive and negative coefficient features
    positive_features = [f for f in sig_feature_names if sig_01[sig_01['Feature Name'] == f]['Coef'].values[0] > 0]
    negative_features = [f for f in sig_feature_names if sig_01[sig_01['Feature Name'] == f]['Coef'].values[0] < 0]
    
    # Create heatmap data
    heatmap_data = pd.DataFrame({
        'Feature': sig_feature_names,
        'Feature Value': sig_values,
        'Coef': [sig_01[sig_01['Feature Name'] == f]['Coef'].values[0] for f in sig_feature_names],
        'Type': ['Malignancy Indicator' if f in positive_features else 'Benign Indicator' for f in sig_feature_names]
    })
    heatmap_data = heatmap_data.sort_values(['Type', 'Coef'], ascending=[True, False])
    
    # Shorten feature names for display
    def shorten_name(name):
        return name.replace('_mean_over_images', ' (mean)').replace('_max_over_images', ' (max)')
    
    heatmap_data['Short Name'] = heatmap_data['Feature'].apply(shorten_name)
    
    # Draw horizontal bar chart
    colors = ['#d73027' if t == 'Malignancy Indicator' else '#4575b4' for t in heatmap_data['Type']]
    bars = ax_heatmap.barh(range(len(heatmap_data)), heatmap_data['Feature Value'], color=colors, alpha=0.7)
    ax_heatmap.set_yticks(range(len(heatmap_data)))
    ax_heatmap.set_yticklabels(heatmap_data['Short Name'], fontsize=9)
    ax_heatmap.axvline(x=0, color='black', linestyle='-', linewidth=0.5)
    ax_heatmap.set_xlabel('Feature Value (Before Standardization)', fontsize=10)
    ax_heatmap.set_title('12 Statistically Significant Feature Values\n(Red=Malignancy Indicator, Blue=Benign Indicator)', fontsize=11)
    
    # Right: Feature contribution (coefficient Ã— standardized feature value)
    ax_contrib = fig.add_subplot(gs_bottom[1])
    
    # Calculate contribution - standardize feature values first
    sig_means = df_nodule_features[sig_feature_names].mean().values
    sig_stds = df_nodule_features[sig_feature_names].std().values
    
    # Reorder according to heatmap_data order
    ordered_indices = [sig_feature_names.index(f) for f in heatmap_data['Feature']]
    ordered_sig_values = [sig_values[i] for i in ordered_indices]
    ordered_means = [sig_means[i] for i in ordered_indices]
    ordered_stds = [sig_stds[i] for i in ordered_indices]
    
    sig_values_scaled = [(v - m) / s if s > 0 else 0 for v, m, s in zip(ordered_sig_values, ordered_means, ordered_stds)]
    contributions = [sv * c for sv, c in zip(sig_values_scaled, heatmap_data['Coef'].values)]
    heatmap_data['Contribution'] = contributions
    
    # Draw contribution bar chart
    colors_contrib = ['#d73027' if c > 0 else '#4575b4' for c in heatmap_data['Contribution']]
    ax_contrib.barh(range(len(heatmap_data)), heatmap_data['Contribution'], color=colors_contrib, alpha=0.7)
    ax_contrib.set_yticks(range(len(heatmap_data)))
    ax_contrib.set_yticklabels([])  # Share y-axis labels
    ax_contrib.axvline(x=0, color='black', linestyle='-', linewidth=0.5)
    ax_contrib.set_xlabel('Contribution to Malignancy Prediction\n(Coef Ã— Standardized Feature Value)', fontsize=10)
    ax_contrib.set_title('Feature Contribution\n(Red>0 towards Malignant, Blue<0 towards Benign)', fontsize=11)
    
    plt.tight_layout()
    plt.show()
    
    # Print detailed information
    print("\n" + "="*70)
    print(f"Nodule Details: {nodule_id}")
    print("="*70)
    print(f"Number of Patches: {n_patches}")
    print(f"True Label: {true_label} ({label_text})")
    print(f"Predicted Probability: {pred_prob:.4f}")
    print(f"Predicted Label: {pred_label} ({pred_text})")
    print(f"Prediction Result: {'Correct âœ“' if pred_label == true_label else 'Incorrect âœ—'}")
    
    return heatmap_data

# Select a nodule for display - select a malignant nodule
malignant_nodules = df_nodule_features[df_nodule_features['malignancy_label'] == 1]['pid_nid_combo'].tolist()
example_nodule = malignant_nodules[10]  # Select the 11th malignant nodule

print(f"Selected nodule: {example_nodule}")
result_df = case_study_nodule(
    nodule_id=example_nodule,
    df_image_features_withmeta=df_image_features_withmeta,
    df_nodule_features=df_nodule_features,
    lasso_model=lasso_model,
    scaler_full=scaler_full,
    X_cols=X_cols,
    sig_feature_names=sig_feature_names,
    sig_01=sig_01
)
```
